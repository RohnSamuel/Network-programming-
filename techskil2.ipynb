```python
#1data cleaning + handling missing values
import pandas as pd
import numpy as np
data={
    "Name":["Kumar", "Meena","Ravi","Priya"],
    "Age":[25,None,22,None],
    "Marks":[88,92,None,75]
}
df=pd.DataFrame(data)
print("Original Data:")
print(df)
#fill missing ages with mean age
df["Age"]=df["Age"].fillna(df["Age"].mean())

#fill missing marks with median marks
df["Marks"]=df["Marks"].fillna(df["Marks"].median())

print("\nAfter Cleaning:")
print(df)

```

```python
#2label encoding-converting categories to numbers
from sklearn.preprocessing import LabelEncoder
import pandas as pd
df=pd.DataFrame({
    "Name":['Kumar','Meena','Ravi','Kumar'],
    "City":['Chennai','Delhi','Chennai','Mumbai']

})
print("Original:")
print(df)
le=LabelEncoder()
df['City_Code']=le.fit_transform(df['City'])
print("\nAfter Encoding:")
print(df)
print("\nMapping",dict(zip(le.classes_,le.transform(le.classes_))))
```

```python
#feature selection using correlation
import pandas as pd
df=pd.DataFrame({
    "StudyHours":[1,2,3,4,5],
    "Attendance":[60,65,70,80,95],
    "Marks":[50,55,65,75,90]
})
print(df.corr())
```

```python
#4train-test split with X and Y values
from sklearn.model_selection import train_test_split
import pandas as pd
df=pd.DataFrame({
    "Hours":[2,3,4,5,6,7],
    "Marks":[40,50,60,70,80,90]
})
X=df[["Hours"]]
y=df["Marks"]
X_train, X_test, y_train, y_test=train_test_split(
    X,y,test_size=0.3, random_state=42
)
print("X_train:\n", X_train)
print("\ny_train:\n",y_train)
print("\nX_test:\n",X_test)
print("\ny_test:\n",y_test)
```

```python
#5 simple linear regresion
from sklearn.linear_model import LinearRegression
import pandas as pd
df=pd.DataFrame({
    "Hours":[1,2,3,4,5],
    "Marks":[35,45,50,70,80]
})
X=df[["Hours"]]
y=df["Marks"]
model=LinearRegression()
model.fit(X,y)

print("Slope:",model.coef_[0])
print("Intercept:",model.intercept_)
print("Predicted Marks for 6 hours:",model.predict([[6]])[0])
```
